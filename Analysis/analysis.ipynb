### Null values :

```python
nulls = df.isnull().sum()
nulls
```

```python
nulls_df = nulls.reset_index()
nulls_df.columns = ['Column', 'Null_Count']

plt.figure(figsize=(8, 8))
sns.barplot(x='Null_Count', y='Column', data=nulls_df, palette='pastel')
plt.title('Number of Null Values in Each Column')
plt.xlabel('Number of Null Values')
plt.ylabel('Columns')
plt.show()
```

```python
def fill_mean_na(mean):
    for mn in mean:
        df[mn].fillna(df[mn].mean(),inplace=True)
def fill_mode_na(mode):
    for md in mode:
        df[md].fillna(df[md].mode()[0],inplace=True)
def fill_unknown_na(unknown):
    for un in unknown:
        df[un].fillna("unknown",inplace=True)
```

```python
df["Age"].fillna(round(df['Age'].mean()),inplace=True)
df["EmployeeCount"].fillna(1,inplace=True)
mean = ["DailyRate","DistanceFromHome","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager","PercentSalaryHike","StandardHours","TotalWorkingYears","HourlyRate","MonthlyRate","MonthlyIncome","NumCompaniesWorked"]
fill_mean_na(mean)
mode = ["WorkLifeBalance","PerformanceRating","RelationshipSatisfaction","StockOptionLevel","Education","EnvironmentSatisfaction","EnvironmentSatisfaction","JobInvolvement","JobLevel","JobSatisfaction",]
fill_mode_na(mode)
unknown = ["Department","EducationField","Employee Source","JobRole","BusinessTravel","MaritalStatus","Over18","OverTime","Gender"]
fill_unknown_na(unknown)

nulls_cnt = df.isnull().sum().sum()
if(nulls_cnt==0):
    print("No null values")
else:
    print("There are "+str(nulls_cnt)+" null values to be removed")
```

```python
df = df.dropna()
```

```python
nulls_cnt = df.isnull().sum().sum()
if(nulls_cnt==0):
    print("No null values")
else:
    print("There are "+str(nulls_cnt)+" null values to be removed")
```

#### Correlation Matrix:
Relation between features 

```python
plt.figure(figsize=(14, 12))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

From here we can see the relation between the **MonthlyIncome** and **JobLevel** have close relation so we can use any one column,instead of both

#### Age vs. Attrition

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Attrition', y='Age', data=df, inner='quartile')
plt.title('Age vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
```

#### Monthly Income vs. Attrition

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Attrition', y='MonthlyIncome', data=df, inner='quartile')
plt.title('Monthly Income vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

#### Department vs. Attrition

```python
plt.figure(figsize=(10, 6))
department_attrition = pd.crosstab(df['Department'], df['Attrition'])
sns.heatmap(department_attrition, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Department vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Department')
plt.show()
```

###  JobRole vs. Attrition

```python
plt.figure(figsize=(12, 6))
jobrole_attrition = pd.crosstab(df['JobRole'], df['Attrition'])
sns.heatmap(jobrole_attrition, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Job Role vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Job Role')
plt.show()
```

**Sales Executives** and **Research Scientists** have the highest number of current employees, attrition, and voluntary resignations among all job roles.

### Average Monthly Income vs. Attrition


```python
plt.figure(figsize=(8, 6))
avg_monthly_income = df.groupby('Attrition')['MonthlyIncome'].mean().reset_index()
sns.barplot(x='Attrition', y='MonthlyIncome', data=avg_monthly_income)
plt.title('Average Monthly Income vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Average Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
avg_total_working_years = df.groupby('Attrition')['TotalWorkingYears'].mean().reset_index()
sns.barplot(x='Attrition', y='TotalWorkingYears', data=avg_total_working_years)
plt.title('Average Total Working Years vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Average Total Working Years')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.title('Business Travel vs. Attrition')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title('Job Role vs. Attrition')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Sales Executives** and **Research Scientists** have high attrition rates, whereas **Managers** and **Manufacturing Directors** have lower voluntary resignations.
